```python
import re
import numpy as np
```

```python
years = []
for item in data['time']:
    year=re.findall(r"\d{4}", item)[0]
    years.append(int(year))
year = np.array(years)
data['year'] = year
```

```python
plt.figure(figsize=(10,2))
data['year'].plot.hist(grid=True)
#plt.xticks(np.arange(1970, 2021,1),rotation=45)
plt.grid(axis='y', alpha=0.75)
```

```python
y = list(range(1970,2020))
plt.figure(figsize=(20,2))
plt.xlim((1970,2020))
plt.bar(y, [list(data['year']).count(i) for i in y], color='g')
plt.xticks(np.arange(1970,2020,1), rotation=45)
plt.show()
```
